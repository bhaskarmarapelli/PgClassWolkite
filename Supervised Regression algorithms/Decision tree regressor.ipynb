# decision tree Regression

```python
import pandas as pd

import os
os.chdir("C:\\Users\\bhaskar\\python practice\\datasets")
bikes = pd.read_csv('bike.csv')

bikes.head()
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(8,6))
plt.plot(bikes['temp'], bikes['count'], 'o')
plt.xlabel('temperature')
plt.ylabel('bikes')
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor

import numpy as np
regressor = DecisionTreeRegressor(max_depth=2)
regressor.fit(np.array([bikes['temp']]).T, bikes['count'])
```

```python
xx = np.array([np.linspace(-5, 40, 100)]).T
plt.figure(figsize=(8,6))
plt.plot(bikes['temp'], bikes['count'], 'o', label='observation')
plt.plot(xx, regressor.predict(xx), linewidth=4, alpha=.7, label='prediction')
plt.xlabel('temperature')
plt.ylabel('bikes')
plt.legend()
plt.show()
```

```python
from sklearn.tree import export_graphviz

export_graphviz(regressor, out_file='tree.dot', feature_names=['temp'])
```

```python
from sklearn import tree
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(regressor,feature_names=['temp'],filled=True)
```
